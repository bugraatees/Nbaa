# NBA DATASET

* Below I will explain basic analyzes and step-by-step data visualizations you can do on an NBA dataset dowloaded from Kaggle. 
While performing these analyses, we will use popular libraries such as Python, Pandas, Matplotlib and Seaborn

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("nba.csv")
```

```python
df.head()
```

```python
#Checking NaN values
df.isnull().sum()
```

```python
#Remove rows with NaN values
df = df.dropna()
```

```python
#Number of players by team
plt.figure(figsize=(14 , 8))
df["Team"].value_counts().plot(kind = 'bar' , color = "blue")
plt.title("Number of Players in Each NBA Team")
plt.xlabel("Team")
plt.ylabel("Number Of Players")
plt.show()
```

```python
#Salary distribution
plt.figure(figsize=(8,6))
sns.histplot(df["Salary"].dropna(), bins=30, kde = True, color="purple")
plt.title("Salary Distribution of NBA Players")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
#Salary by teams
team_salary = df.groupby("Team")["Salary"].mean().sort_values(ascending = False)
plt.figure(figsize=(12,8))
team_salary = sns.scatterplot(x='Team', y='Salary', data=df, color='red')
plt.title('Average Salary by NBA Team')
plt.xlabel('Team')
plt.ylabel('Average Salary')
plt.xticks(rotation=90)
plt.show()
```

# Result : 


The analyzes provide detailed information and allow you to determine your road map accordingly:

* Number of team players: Although the number of players in teams is generally equal, some minor differences can be observed. These differences may be related to coaches' rotation strategies, injuries, or other tactical decisions.

* 
Salary Breakdown: There is a huge gap in salaries among NBA players. While star players receive big contracts, other players often play for lower salaries. This shows that the salary structure in the NBA is "star-oriented".

* Salary Differences by Teams: The higher average salaries of some teams indicate that these teams have more high-paid star players. Salary averages may vary depending on the squad structure and financial flexibility of the teams.

These results provide information about how NBA player salaries and team structures are shaped, player compensation and team roster strategies.
